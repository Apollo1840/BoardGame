```python
import pandas as pd
import json
```

```python
df = pd.read_csv(
    'monster_cards.csv',
    sep='|',
    encoding='utf-8',
    converters={
        'characters': lambda x: json.loads(x) if x and x != '{}' else {},
        'skills': lambda x: json.loads(x) if x and x != '[]' else []
    }
)
```

```python
# Insert an empty 'effect' column after 'defence' and save to CSV
cols = list(df.columns)
effect_idx = cols.index('defence') + 1
cols = cols[:effect_idx] + ['effect'] + cols[effect_idx:]
df['effect'] = ''
df = df[cols]
df.to_csv('monster_cards.csv', sep='|', index=False, encoding='utf-8')
df.head()
```
